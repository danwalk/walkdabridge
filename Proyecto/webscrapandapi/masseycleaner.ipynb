```python
import pandas as pd
import numpy as np
import json
import requests
```

```python
with open('massey.json') as json_file:
    mas = json.load(json_file)
```

```python
mas = pd.DataFrame.from_dict(mas, orient='index')
```

```python
mas
```

```python
mas[['AwayScore','HomeScore']] = mas.Score.str.split("\n",expand=True)
mas.drop(columns=['Score'], inplace=True)
mas
```

```python
mas[['AwayWinPred','HomeWinPred']] = mas.w.str.split("\n",expand=True)
mas.drop(columns=['w'], inplace=True)
mas
```

```python
mas[['AwayScorePred','HomeScorePred']] = mas.Prediction.str.split("\n",expand=True)
mas.drop(columns=['Prediction'], inplace=True)
mas
```

```python
mas[['AwayTeam','HomeTeam']] = mas.Teams.str.split("\n@ ",expand=True)
#mas.drop(columns=['Prediction'], inplace=True)
mas
```

```python
mas.drop(columns=['Teams'], inplace=True)
mas
```

```python
mas[['Date','h']] = mas.G.str.split("\n",expand=True)
#mas.drop(columns=['Prediction'], inplace=True)
mas
```

```python
mas.drop(columns=['h'], inplace=True)
mas
```

```python
mas.to_csv("cleanmass.csv")
```
