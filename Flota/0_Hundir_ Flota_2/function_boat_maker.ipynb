```python
#los generadores de barcos para cada direccion


def genwest(boatsize, c, r):
    positionlist = [1 for i in range((boatsize-1))]
    shipcoord = []
    shipcoord.append(c+r)
    for x in positionlist:
        c = int(c)-x
        shipcoord.append(str(c)+r)
    return shipcoord

def geneast(boatsize, c, r):
    positionlist = [1 for i in range((boatsize-1))]
    shipcoord = []
    shipcoord.append(c+r)
    for x in positionlist:
        c = int(c)+x
        shipcoord.append(str(c)+r)
    return shipcoord

def gennorth(boatsize, c, r):
    dictwe = {"a":1, "b":2, "c":3, "d":4, "e":5, "f":6, "g":7, "h":8, "i":9, "j":10}
    dictns = {1:"a", 2:"b", 3:"c", 4:"d", 5:"e", 6:"f", 7:"g", 8:"h", 9:"i", 10:"j"}
    positionlist = [1 for i in range((boatsize-1))]
    shipcoord = []
    shipcoord.append(c+r)
    n = dictwe[r]
    for x in positionlist:
        n-=1
        g = dictns[n]
        shipcoord.append(str(c)+g)
    return shipcoord

def gensouth(boatsize, c, r):
    dictwe = {"a":1, "b":2, "c":3, "d":4, "e":5, "f":6, "g":7, "h":8, "i":9, "j":10}
    dictns = {1:"a", 2:"b", 3:"c", 4:"d", 5:"e", 6:"f", 7:"g", 8:"h", 9:"i", 10:"j"}
    positionlist = [1 for i in range((boatsize-1))]
    shipcoord = []
    shipcoord.append(c+r)
    n = dictwe[r]
    for x in positionlist:
        n+=1
        g = dictns[n]
        shipcoord.append(str(c)+g)
    return shipcoord
```
